# Funciones algebraicas

```python
import matplotlib.pyplot as plt 
import numpy as np
```

## Función lineal

```python
def f(m, x, b):
        return m * x + b

res = 100
m = -2
b= 5
x = np.linspace(-10, 10, num=res)
y = f(m, x, b)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

## Función Polinómica

```python
def f(x):
        return 2*x**8 - 3*x**2 + 4

res = 100
x = np.linspace(-10, 10, num=res)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

## Funciones trascentendes

```python
def f(x):
        return np.sin(x)

res = 100
x = np.linspace(-10, 10, num=res)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

## Función exponencial

```python
def f(x):
        return np.e**x

res = 100
x = np.linspace(-1, 1, num=res)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

## Función logaritmo

```python
def f(x):
        return np.log2(x)

res = 1000
x = np.linspace(0.01, 256, num=res)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

## Función seccionada

```python
def H(x):
    y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x >= 0:
            y[idx] = 1
    return y


res = 1000
x = np.linspace(-10, 10, num=res)
y = H(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
# ax.axhline(y=0, color='r')
# ax.axvline(x=0, color='r')
plt.show()
```
